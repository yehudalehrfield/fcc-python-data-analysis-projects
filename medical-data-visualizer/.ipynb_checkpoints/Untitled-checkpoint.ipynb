Normalize Data: 0 is *good* and 1 is *bad*
* cholesterol: change all 1's to 0's, anything above 1 to 1
* glucose: change all 1's to 0's, anything above 1 to 1

```python
#df_normalized = df.copy()
```

```python
#(df_normalized['cholesterol'] > 1).astype(int)
```

```python
# df_normalized['cholesterol'] = (df_normalized['cholesterol'] > 1).astype(int)
df['cholesterol'] = (df['cholesterol'] > 1).astype(int)
```

```python
# (df_normalized['gluc'] > 1).astype(int)
(df['gluc'] > 1).astype(int)
```

```python
# df_normalized['gluc'] = (df_normalized['gluc'] > 1).astype(int)
df['gluc'] = (df['gluc'] > 1).astype(int)
```
